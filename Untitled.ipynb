# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtypes=='object':
        df[col] = le.fit_transform(df[col])
```

# Data Split Into Training and Testing 

```python
X = df.iloc[:,0:12]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Model Buliding

Logistic Regression


```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred1 = lr.predict(X_test)
print("Accuracy Logistic R: ", accuracy_score(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print("Precision Logistic R: ",precision_score(y_test, y_pred1))
```

Random Forest

```python
random_f = RandomForestClassifier()
random_f.fit(X_train, y_train)
y_pred2 = random_f.predict(X_test)
print("Accuracy Random F: ", accuracy_score(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
print("Precision Random F: ",precision_score(y_test, y_pred2))
```

Adaboost Classifier

```python
adaboost = AdaBoostClassifier()
adaboost.fit(X_train, y_train)
y_pred3 = adaboost.predict(X_test)
print("Accuracy Adaboost C: ", accuracy_score(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
print("Precision Adaboost C: ",precision_score(y_test, y_pred3))
```

Decision Tree

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred4 = tree.predict(X_test)
print("A Decision T: ", accuracy_score(y_test, y_pred4))
print(confusion_matrix(y_test, y_pred4))
print("Precision Decision Tree: ",precision_score(y_test, y_pred4))
```

# Classification Report

```python
logistic_report = classification_report(y_test, y_pred1)
random_report = classification_report(y_test, y_pred2)
adaboost_report = classification_report(y_test, y_pred3)
tree_report = classification_report(y_test, y_pred4)
```

```python
print('\n Logistic Regression \n', logistic_report)
print()
print('\n Random Forest Classifier\n', random_report)
print()
print('\n Adaboost Classifier\n', adaboost_report)
print()
print('\n Decision Tree Classifier\n', tree_report)
print()
```

```python
result=pd.DataFrame({'Classification Algorithms':['Logistic Regression','Random Forest Classifier','AdaBoost Classifier', 'Decision Tree'],
                  ' Test Accuracy':[accuracy_score(y_test,y_pred1),accuracy_score(y_test,y_pred2),accuracy_score(y_test,y_pred3), accuracy_score(y_test,y_pred4)],
                  ' Precision Score':[precision_score(y_test,y_pred1, average='macro'),precision_score(y_test,y_pred2,average='macro'),precision_score(y_test,y_pred3,average='macro'),
                                    precision_score(y_test,y_pred4,average='macro')]
                  
})
result
```
